### How to Use Preattentive Attributes to Emphasize the Last Three Months

- **Color**  
  Highlight the last three months in a bold or contrasting color (e.g., red or blue), while rendering the rest of the line in a muted gray.

- **Line Thickness**  
  Make the line thicker for the last three months compared to the rest of the series.

- **Markers**  
  Add distinct markers (e.g., circles or squares) only to the last three data points.

- **Annotations**  
  Label the last three months with their values or brief notes (e.g., “Recent spike”).

- **Motion or Animation**  
  If the graph is interactive, animate the drawing of the last three months after the rest of the line is drawn.

- **Blur or Desaturate Earlier Data**  
  Fade out or blur the earlier months to reduce their visual weight and draw attention to the most recent data.

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
# Page 118 line graph

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate monthly data
months = pd.date_range(start='2023-01-01', periods=12, freq='M').strftime('%b')

# Generate two closely related lines with some noise
np.random.seed(0)
line1 = 1000 * (np.random.normal(size=12) / 5 + np.arange(12) / 16)
line2 = line1 + np.random.normal(size=12) * 100

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(months, line1, marker='o', color='blue', label='Brand A')
plt.plot(months, line2, marker='o', color='green', label='Brand B')

# Add data labels for the last 3 months
for i in range(9, 12):
    plt.text(months[i], line1[i], f'{line1[i]:.2f}', color='blue', ha='center', va='bottom')
    plt.text(months[i], line2[i], f'{line2[i]:.2f}', color='green', ha='center', va='bottom')

# Customize the plot
plt.title('Monthly Sales Volume')
plt.xlabel('Month')
plt.ylabel('Sales Volume')
plt.legend()
plt.grid(False)
plt.show()

```
