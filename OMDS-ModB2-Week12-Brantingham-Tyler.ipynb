# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

## Graph 1
### Advantages
- Show different y-axis scales so you an fairly compare trends of different magnitudes.
- Less overlap/distraction
### Disadvantages
- Difficult to compare as they're on different scales
- It's impossible to differentiate which each set represents due to lack of color/legend/labelling

## Graph 2
### Advantages
- Easy to differentiate each set by color
- Fairly compare on the same y-axis
### Disadvantages
- Some overlapping data can be distracting
- If lines were close to each other it would be difficult to see.

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 187-205 of the Storytelling With Data book as best you can. (Chapter Eight). You do not have to get the exact data values right, just the overall look and feel.

```python
data = {
    'Product': ['Product A', 'Product A', 'Product A', 'Product A', 'Product A', 'Product A', 'Product A',
                'Product B', 'Product B', 'Product B', 'Product B', 'Product B', 'Product B', 'Product B',
                'Product C', 'Product C', 'Product C', 'Product C', 'Product C',
                'Product D', 'Product D', 'Product D', 'Product D',
                'Product E', 'Product E'],
    'Year': [2008, 2009, 2010, 2011, 2012, 2013, 2014,
             2008, 2009, 2010, 2011, 2012, 2013, 2014,
             2010, 2011, 2012, 2013, 2014,
             2011, 2012, 2013, 2014,
             2013, 2014],
    'Price': [350, 325, 300, 275, 250, 225, 200,
              375, 350, 325, 300, 275, 250, 225,
              150, 125, 100, 175, 150,
              200, 175, 150, 125,
              50, 25]
}

df = pd.DataFrame(data)

fig, ax = plt.subplots(figsize=(12, 8))
bar_width = 0.12
products = df['Product'].unique()
years = df['Year'].unique()
x = np.arange(len(products))

colors = ['#4e79a7', '#f28e2b', '#e15759', '#76b7b2', '#59a14f', '#edc948', '#b07aa1']
year_colors = {year: colors[i] for i, year in enumerate(years)}

for i, year in enumerate(years):
    prices = []
    for product in products:
        price_row = df[(df['Product'] == product) & (df['Year'] == year)]
        if not price_row.empty:
            prices.append(price_row['Price'].values[0])
        else:
            prices.append(0)  # Use 0 for missing data

    x_positions = x + i * bar_width - (len(years) / 2 - 0.5) * bar_width
    bars = ax.bar(x_positions, prices, bar_width, label=str(year), color=year_colors.get(year))

ax.set_title('Average Retail Product Price Per Year', fontsize=16, pad=20)
ax.set_ylabel('$', fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(products, fontsize=12)
ax.set_yticks(np.arange(0, 501, 100))
ax.set_yticklabels(['$0', '$100', '$200', '$300', '$400', '$500'], fontsize=12)
ax.set_ylim(0, 500)
ax.legend(title='Year', loc='center left', bbox_to_anchor=(1, 0.5))
ax.grid(axis='y', linestyle='-', alpha=0.5)

ax.tick_params(axis='x', which='both', bottom=True, top=False, labelbottom=True)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
```
